```python
X = gov[[
    'comments_count_fb',
    'reactions_count_fb',
    'shares_count_fb',
    'engagement_fb',
    'comment_likes_count'
]]
kmeans = KMeans(
    n_clusters = 8, # testar diferentes valores
    n_init = 15, 
    max_iter = 450, 
    init = 'k-means++'
)
kmeans.fit(X)

gov['cluster'] = kmeans.labels_ # adiciona coluna c/ cluster
grouped = gov.groupby('cluster')
stacked = pd.DataFrame()
for group in grouped:
    means = group[1].mean(axis = 0)
    
    stacked[group[0]] = means
print(stacked)
```

```python
silhouette_avg = silhouette_score(X, kmeans.labels_)
print('silhueta media:', silhouette_avg)
```

```python
import matplotlib.pyplot as plt
 
plt.scatter(X['shares_count_fb'], X['engagement_fb'], s = 100, c = kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'red',label = 'Centroids')
plt.title('Engajamento Cluster')
plt.xlabel('Engajamento')
plt.ylabel('Compartilhamento')
plt.legend()
 
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt
```

```python
# k means determine k
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X)
    kmeanModel.fit(X)
    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0])
 
# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distorção ')
plt.title('Método Elbow Frejat otimização K')
plt.show()
 
```

```python
sentilexpt = open('C:/Users/NB/Desktop/eleitorado/SentiLex-PT01/SentiLex-lem-PT01.txt',  'r')
```

```python
dic_palavra_polaridade = {}
for i in sentilexpt.readlines():
    pos_ponto = i.find('.')
    palavra = (i[:pos_ponto])
    pol_pos = i.find('POL')
    polaridade = (i[pol_pos+4:pol_pos+6]).replace(';','')
    dic_palavra_polaridade[palavra] = polaridade
```

```python
def Score_sentimento(frase):
    frase = frase.lower()
    l_sentimento = []
    for p in frase.split():
        l_sentimento.append(int(dic_palavra_polaridade.get(p, 0)))
    score = sum(l_sentimento)
    if score > 0:
        return 'Positivo, Score:{}'.format(score)
    elif score == 1:
        return 'Neutro, Score:{}'.format(score)
    else:
        return 'Negativo, Score:{}'.format(score)
```
