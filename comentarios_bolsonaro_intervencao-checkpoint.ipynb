```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity
```

```python
leany = pd.read_csv("C:/Users/NB/Desktop/BCK/despesas_parlamentares/data_comentarios_intervencao.csv", delimiter = ';', encoding='iso-8859-1', usecols = ['comment_message'])
```

```python
leany = [e for e in leany['comment_message'] if isinstance(e, str)]
leany = pd.DataFrame(leany)
```

```python
vectorizer = TfidfVectorizer() # variar parametros (L1/L2, etc)
X = vectorizer.fit_transform(leany[0])
X = normalize(X)
```

```python
kmeans = KMeans(
    n_clusters = 3, # variar
    n_init = 15, 
    max_iter = 450, 
    init = 'k-means++'
)
```

```python
kmeans.fit(X)
```

```python
leany['cluster'] = kmeans.labels_
leany
```

```python
text = ['TEMER']
```

```python
X_new = vectorizer.transform(text)
leany['coseno'] = cosine_similarity(X_new, X)[0]
print(leany.sort_values(by = 'coseno'))
```
