```python
def Score_sentimento(frase):
    frase = frase.lower()
    l_sentimento = []
    for p in frase.split():
        l_sentimento.append(int(dic_palavra_polaridade.get(p, 0)))
    score = sum(l_sentimento)
    if score > 0:
        return 'Positivo'
    elif score == 1:
        return 'Neutro'
    else:
        return 'Negativo'
```

```python
sentilexpt = open('C:/Users/NB/Desktop/eleitorado/SentiLex-PT01/SentiLex-lem-PT01.txt',  'r')
```

```python
dic_palavra_polaridade = {}
for i in sentilexpt.readlines():
    pos_ponto = i.find('.')
    palavra = (i[:pos_ponto])
    pol_pos = i.find('POL')
    polaridade = (i[pol_pos+4:pol_pos+6]).replace(';','')
    dic_palavra_polaridade[palavra] = polaridade
```

```python
correio['cat_size2'] = correio['comment_message'].apply(Score_sentimento)
```

```python
correio = [e for e in correio['comment_message'] if isinstance(e, str)]
correio = pd.DataFrame(correio)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity
vectorizer = TfidfVectorizer() # variar parametros (L1/L2, etc)
X = vectorizer.fit_transform(correio[0])
X = normalize(X)
```

```python
kmeans = KMeans(
    n_clusters = 10, # variar
    n_init = 15, 
    max_iter = 450, 
    init = 'k-means++'
)
```

```python
kmeans.fit(X)
```

```python
correio['cluster'] = kmeans.labels_
correio
```
