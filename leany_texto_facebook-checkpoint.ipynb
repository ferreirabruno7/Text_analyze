```python
def answer_three():
    dist = nltk.FreqDist(text1)
    sorted_x = sorted(dist.items(),key=operator.itemgetter(1))
    sorted_x.reverse()
    freqwords = [w for w in sorted_x if len(w) > 5 and dist[w] > 10]
    return sorted_x[:20]

answer_three()
```

```python
def answer_four():
    dist = nltk.FreqDist(text1)
    vocab1 = dist.keys()
    freqwords = [w for w in vocab1 if len(w) > 3 and dist[w] > 15]
    return sorted(freqwords)

answer_four()
```
